### Pull Constellations from Celestrak

```python
# https://celestrak.com/NORAD/elements/master.php
nameList = ['100OrSoBrightest','ActiveGEO','Active','AmateurRadio','Analyst','ARGOS','Beidou','BREEZE-MRBBreakUp','COSMOS2251Debris',
            'CubeSats','DisasterMonitoring','EarthResources','Education','Engineering','Experimental','FENGYUN1CDebris','Galileo','Geodetic',
            'GlobalStar','GLONASSOperational','GOES','Gorizont','GPSOperational','IndianASATTestDebris','Intelsat','Iridium','Iridium33Debris',
            'IridiumNEXT','Last30DaysLaunches','MiscMilitary','Molniya','NNSS','NOAA','Orbcomm','Other','OtherComm','Planet',
            'RadarCalibration','Raduga','RusssianLEONavigation','SBAS','SatNOGs','Search&Rescue','SES','Space&EarthScience','SpaceStations','Spire','Starlink','TDRSS','Weather']

tceList = ['visual','geo','active','amateur','analyst','argos','beidou','2012-044','cosmos-2251-debris',
         'cubesat','dmc','resource','education','engineering','x-comm','1999-025','galileo','geodetic',
         'globalstar','glo-ops','goes','gorizont','gps-ops','2019-006','intelsat','iridium','iridium-33-debris',
         'iridium-NEXT','tle-new','military','molniya','nnss','noaa','orbcomm','other','other-comm','planet',
         'radar','raduga','musson','sbas','satnogs','sarsat','ses','science','stations','spire','starlink','tdrss','weather']

constellations = zip(nameList,tceList)

for name,tce in constellations:
    link = 'https://celestrak.com/NORAD/elements/'+tce+'.txt'
    try:
        f = urllib.request.urlopen(link)
        data = f.read()
        data = data.decode('utf-8').replace('\r','\n')
        data = data.split('\n')
        data = data[0::2]
        file = open(cwdFiles+'\\Constellations\\Celestrak'+name+'.tce','w')
        k = 0
        for line in data:
            if k % 3 != 0:
                file.write(line.replace('\r','')+'\n')
            k += 1
        # file.writelines(data)
        file.close()
        
    except:
        print('Failed To Retrieve: '+tce)
        pass
```
